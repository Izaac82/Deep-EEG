```python
# Import python packages for ML modeling and evaluation

import pandas as pd
import matplotlib.pyplot as plt
import time

from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC

from sklearn import metrics
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn import preprocessing

%matplotlib inline
```

```python
# Load train/test dataset and build subject-trial column

train_dataset = pd.read_csv('../dataset/eeg_train_dataset.csv')
test_dataset = pd.read_csv('../dataset/eeg_test_dataset.csv')

train_dataset['Subject-Trial'] = train_dataset['Subject_id'].astype(str) + '-' + train_dataset['Trial_id'].astype(str)
test_dataset['Subject-Trial'] = test_dataset['Subject_id'].astype(str) + '-' + test_dataset['Trial_id'].astype(str)
```

```python
# Check the dataset

train_dataset.head()
```

```python
# Quantile Aggregation
# Aggregate dataset by column with quantile point values and mean value
# Quantile Points ( min / 0.25 / 0.5 / 0.75 / max ) + Mean Value

def q_aggregation(full_dataframe):
    # check the time
    start_time = time.time()
    
    # list for aggregation
    q_based_aggregation = []
    
    # for each subject-trial, do q-aggregation method
    full_subject_trial = full_dataframe['Subject-Trial'].value_counts().index
    for subject_trial in full_subject_trial:
        split_num = subject_trial.split('-')
        subject = int(split_num[0])
        trial = int(split_num[1])
        single_dataframe = full_dataframe[(full_dataframe['Subject_id'] == subject) & (full_dataframe['Trial_id'] == trial)]

        # get values from dataframe
        alcohol = single_dataframe['Alcohol'].values[0]
        exp = single_dataframe['Exp'].values[0]
        subject_id = single_dataframe['Subject_id'].values[0]
        trial_id = single_dataframe['Trial_id'].values[0]

        # drop the columns which not aggregate to q value
        single_dataframe = single_dataframe.drop(['Time-Series', 'Alcohol', 'Subject_id', 'Exp', 'Trial_id', 'Subject-Trial'], axis=1)

        # calculate q values, then add string columns
        q_min_df = pd.DataFrame(single_dataframe.min().to_dict(), index = [single_dataframe.index.values[-1]])
        q_min_df.columns = [x + '-min' for x in q_min_df.columns]
        
        q_25_df = pd.DataFrame(single_dataframe.quantile(0.25).to_dict(), index = [single_dataframe.index.values[-1]])
        q_25_df.columns = [x + '-25' for x in q_25_df.columns]
        
        q_50_df = pd.DataFrame(single_dataframe.quantile(0.50).to_dict(), index = [single_dataframe.index.values[-1]])
        q_50_df.columns = [x + '-50' for x in q_50_df.columns]
        
        q_mean_df = pd.DataFrame(single_dataframe.mean().to_dict(), index = [single_dataframe.index.values[-1]])
        q_mean_df.columns = [x + '-mean' for x in q_mean_df.columns]
        
        q_75_df = pd.DataFrame(single_dataframe.quantile(0.75).to_dict(), index = [single_dataframe.index.values[-1]])
        q_75_df.columns = [x + '-75' for x in q_75_df.columns]
        
        q_max_df = pd.DataFrame(single_dataframe.max().to_dict(), index = [single_dataframe.index.values[-1]])
        q_max_df.columns = [x + '-max' for x in q_max_df.columns]
        
        q_df = pd.concat([q_min_df, q_25_df, q_50_df, q_mean_df, q_75_df, q_max_df], axis=1)
        
        q_df['Alcohol'] = alcohol
        q_df['Exp'] = exp
        q_df['Subject_id'] = subject_id
        q_df['Trial_id'] = trial_id

        # append to full data list
        q_based_aggregation.append(q_df)
        print(str(subject) + '-' + str(trial), end=', ')

    # list to one dataframe
    q_based_aggregation = pd.concat(q_based_aggregation)
    print("--- %s seconds ---" % (time.time() - start_time))
    return q_based_aggregation
```

```python
train_dataset_q = q_aggregation(train_dataset)
test_dataset_q = q_aggregation(test_dataset)
```

```python
# Check the aggregated dataset, and every record have different Subject and Trial id

train_dataset_q.head()
```

```python
# Build X feature data and y target data for each train/test dataset

X_train = train_dataset_q.drop(['Alcohol', 'Exp', 'Subject_id', 'Trial_id'], axis=1)
y_train = train_dataset_q['Alcohol']

X_test = test_dataset_q.drop(['Alcohol', 'Exp', 'Subject_id', 'Trial_id'], axis=1)
y_test = test_dataset_q['Alcohol']
```

```python
# Shuffle data

X_train = X_train.sample(frac=1, random_state=42).reset_index(drop=True)
y_train = y_train.sample(frac=1, random_state=42).reset_index(drop=True)
X_test = X_test.sample(frac=1, random_state=42).reset_index(drop=True)
y_test = y_test.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
# Feature scaling

X_train = preprocessing.scale(X_train)
X_test = preprocessing.scale(X_test)
```

```python
# Logistic Regression

def logr(X_train, y_train, X_test, y_test):
    logr = LogisticRegression(solver='liblinear', penalty='l2')
    logr.fit(X_train,y_train)
    y_pred = logr.predict(X_test)
    accuracy = metrics.accuracy_score(y_pred, y_test)
    return accuracy
```

```python
# Gaussian Naive Bayes

def gaussian_nb(X_train, y_train, X_test, y_test):
    gaussian_nb = GaussianNB()
    gaussian_nb.fit(X_train, y_train)
    y_pred = gaussian_nb.predict(X_test)
    accuracy = metrics.accuracy_score(y_pred, y_test)
    return accuracy
```

```python
# K-Nearest Neighborhood Classifier

def knn(X_train, y_train, X_test, y_test):
    knn = KNeighborsClassifier()
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    accuracy = metrics.accuracy_score(y_pred, y_test)
    return accuracy
```

```python
# Decision Tree

def decision_tree(X_train, y_train, X_test, y_test):
    d_tree = tree.DecisionTreeClassifier()
    d_tree.fit(X_train, y_train)
    y_pred = d_tree.predict(X_test)
    accuracy = metrics.accuracy_score(y_pred, y_test)
    return accuracy
```

```python
# Random Forest

def random_forest(X_train, y_train, X_test, y_test):
    rfc = RandomForestClassifier()
    rfc.fit(X_train, y_train)
    y_pred = rfc.predict(X_test)
    accuracy = metrics.accuracy_score(y_pred, y_test)
    return accuracy
```

```python
# Support Vector Machine Classifier with Linear Kernel

def svc_lin(X_train, y_train, X_test, y_test):
    svc = SVC(gamma='auto', kernel='linear')
    svc.fit(X_train, y_train)
    y_pred = svc.predict(X_test)
    accuracy = metrics.accuracy_score(y_pred, y_test)
    return accuracy
```

```python
# Support Vector Machine Classifier with RBF Kernel

def svc_rbf(X_train, y_train, X_test, y_test):
    svc = SVC(gamma='auto', kernel='rbf')
    svc.fit(X_train, y_train)
    y_pred = svc.predict(X_test)
    accuracy = metrics.accuracy_score(y_pred, y_test)
    return accuracy
```

```python
# Support Vector Machine Classifier with Polinomial Kernel

def svc_poly(X_train, y_train, X_test, y_test):
    svc = SVC(gamma='auto', kernel='poly')
    svc.fit(X_train, y_train)
    y_pred = svc.predict(X_test)
    accuracy = metrics.accuracy_score(y_pred, y_test)
    return accuracy
```

```python
# 10-Fold Cross Validation with each models

logr_acc = []
gaussian_nb_acc = []
knn_acc = []
d_tree_acc =[]
svc_lin_acc =[]
svc_rbf_acc =[]
svc_poly_acc =[]

kf = KFold(n_splits=10)
for train_index, val_index in kf.split(X_train):
    X_val_train, X_val_test = X_train[train_index], X_train[val_index]
    y_val_train, y_val_test = y_train[train_index], y_train[val_index]
    
    logr_acc.append(logr(X_val_train, y_val_train, X_val_test, y_val_test))
    gaussian_nb_acc.append(gaussian_nb(X_val_train, y_val_train, X_val_test, y_val_test))
    knn_acc.append(knn(X_val_train, y_val_train, X_val_test, y_val_test))
    d_tree_acc.append(decision_tree(X_val_train, y_val_train, X_val_test, y_val_test))
    svc_lin_acc.append(svc_lin(X_val_train, y_val_train, X_val_test, y_val_test))
    svc_rbf_acc.append(svc_rbf(X_val_train, y_val_train, X_val_test, y_val_test))
    svc_poly_acc.append(svc_poly(X_val_train, y_val_train, X_val_test, y_val_test))
```

```python
# Test accuracy for each models

test_acc = []
test_acc.append(gaussian_nb(X_train, y_train, X_test, y_test))
test_acc.append(decision_tree(X_train, y_train, X_test, y_test))
test_acc.append(knn(X_train, y_train, X_test, y_test))
test_acc.append(logr(X_train, y_train, X_test, y_test))
test_acc.append(svc_lin(X_train, y_train, X_test, y_test))
test_acc.append(svc_rbf(X_train, y_train, X_test, y_test))
test_acc.append(svc_poly(X_train, y_train, X_test, y_test))
```

```python
# Draw box-plot with validation results
# Then, draw red-dot point with test results

plt.rcParams.update({'font.size': 15})
names = ['Gaussian NB', 'Decision Tree', 'KNN', 'LogR', 'SVC_lin', 'SVC_rbf', 'SVC_poly']
data = [gaussian_nb_acc, d_tree_acc, knn_acc, logr_acc, svc_lin_acc, svc_rbf_acc, svc_poly_acc]
x_value = [1, 2, 3, 4, 5, 6, 7]

plt.figure(figsize=(15, 6))
plt.ylim(0.5, 1.0)
plt.boxplot(data)
plt.xticks(x_value, names)
plt.scatter(x_value, test_acc, c='red')

plt.show()
```
